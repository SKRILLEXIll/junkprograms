```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
iris_d =  pd.read_csv("/content/preprocessed.csv")
```

```python
iris_d.head()
```

```python
iris_d.columns
```

```python
iris_d.shape
```

```python
plt.scatter(iris_d['Price'],iris_d['Duration_hours'], color='red')
plt.title("scatter plot")
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.show()
```

```python
plt.hist(iris_d['Duration_hours'], bins=20)
plt.title("Histogram")
plt.xlabel("Sepal width")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(iris_d['Total_Stops'], bins=10)
plt.title("Histogram")
plt.xlabel("Petal width")
plt.ylabel("Frequency")
plt.show()
```

```python
sb.boxplot(x="Duration_hours", data=iris_d)
plt.title("Box Plot")
```
